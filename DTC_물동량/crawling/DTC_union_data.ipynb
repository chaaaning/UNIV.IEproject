```python
import pandas as pd
```

```python
# 크롤링 된 데이터를 엑셀에서 불러오기

df06 = pd.read_excel('06물동량.xlsx')
df07 = pd.read_excel('07물동량.xlsx')
df08 = pd.read_excel('08물동량.xlsx')
df09 = pd.read_excel('09물동량.xlsx')
df10 = pd.read_excel('10물동량.xlsx')
df11 = pd.read_excel('11물동량.xlsx')
df12 = pd.read_excel('12물동량.xlsx')
```

```python
# 품목 갯수 확인

print("06월의 품목 갯수 : ", len(df06))
print("07월의 품목 갯수 : ", len(df07))
print("08월의 품목 갯수 : ", len(df08))
print("09월의 품목 갯수 : ", len(df09))
print("10월의 품목 갯수 : ", len(df10))
print("11월의 품목 갯수 : ", len(df11))
print("12월의 품목 갯수 : ", len(df12))
```

```python
# 불필요 인덱스 삭제

df06 = df06.drop('순번', axis=1)
df06 = df06.drop('Unnamed: 0', axis=1)

df07 = df07.drop('순번', axis=1)
df07 = df07.drop('Unnamed: 0', axis=1)

df08 = df08.drop('순번', axis=1)
df08 = df08.drop('Unnamed: 0', axis=1)

df09 = df09.drop('순번', axis=1)
df09 = df09.drop('Unnamed: 0', axis=1)

df10 = df10.drop('순번', axis=1)
df10 = df10.drop('Unnamed: 0', axis=1)

df11 = df11.drop('순번', axis=1)
df11 = df11.drop('Unnamed: 0', axis=1)

df12 = df12.drop('순번', axis=1)
df12 = df12.drop('Unnamed: 0', axis=1)

df12.head()
```

```python
# Index 기준의 Concat을 사용하기 위해 HS 코드를 새로운 인덱스로 배정

df06 = df06.set_index('코드')
df07 = df07.set_index('코드')
df08 = df08.set_index('코드')
df09 = df09.set_index('코드')
df10 = df10.set_index('코드')
df11 = df11.set_index('코드')
df12 = df12.set_index('코드')
```

```python
# 12월부터 물동량 데이터 취합

df_r = pd.concat([df12['품목명'],df12['12물동량'],
                  df11['11물동량'],df10['10물동량'],
                  df09['09물동량'],df08['08물동량'],
                  df07['07물동량'],df06['06물동량']],axis=1)
df_r
```

```python
df_r.to_excel('인천항물동량종합_역순.xlsx')
```

```python
# 역순 저장된 데이터 조작 후 불러오기 (sum과 average 추가)

df_handle = pd.read_excel('인천항물동량종합.xlsx')
df_handle
```

```python
# 데이터 형 변환 Str -> Float

df_handle['종합물동량'] = df_handle['종합물동량'].astype(float)
df_handle['월평균 물동량'] = df_handle['월평균 물동량'].astype(float)
df_handle
```

```python
# 월평균 물동량으로 상위 5품목 선정

df_handle.sort_values(by='월평균 물동량', ascending=False).head()
```
